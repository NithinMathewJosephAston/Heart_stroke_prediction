```python
import datetime
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
warnings.simplefilter('ignore') # So that we can ignore depreciation issues in Python.
```

```python
def naivebayes_gaussian_model(train_x_1, test_x_1, train_y_1, test_y_1):
    '''
    :param train_x_1: 80% of the train data from the observation variables
    :param test_x_1: 20% of the test data from the observation variables
    :param train_y_1: 80% of the train data from the response variables
    :param test_y_1: 20% of the test data from the response variables
    :return: Nil
    '''
    model_1 = GaussianNB()
    model_1.fit(train_x_1, train_y_1)
    predict_y_1 = model_1.predict(test_x_1)
    model_1_train_score = model_1.score(train_x_1, train_y_1)
    model_1_test_score = model_1.score(test_x_1, test_y_1)
#     print("Train score: {}".format(model_1_train_score))
#     print("Test score: {}".format(model_1_test_score))
    print("Naive Bayes Gaussian model\n")
    print("Train accuracy:", np.round(accuracy_score(train_y_1, model_1.predict(train_x_1)), 2))
    print("Test accuracy:", np.round(accuracy_score(test_y_1, predict_y_1), 2),"\n")
    print("\n Confusion Matrix:\n", pd.crosstab(test_y_1, predict_y_1))
    print("\n Classification Report:\n",classification_report(test_y_1, predict_y_1))
#     return predict_y_1, model_1_train_score, model_1_test_score, model_1


def naivebayes_multinomial_model(train_x_1_1, test_x_1_1, train_y_1_1, test_y_1_1):
    '''
    :param train_x_1_1: 80% of the train data from the observation variables
    :param test_x_1_1: 20% of the test data from the observation variables
    :param train_y_1_1: 80% of the train data from the response variables
    :param test_y_1_1: 20% of the test data from the response variables
    :return: Nil
    '''
    model_1_1 = MultinomialNB()
    model_1_1.fit(train_x_1_1, train_y_1_1)
    predict_y_1_1 = model_1_1.predict(test_x_1_1)
    model_1_1_train_score = model_1_1.score(train_x_1_1, train_y_1_1)
    model_1_1_test_score = model_1_1.score(test_x_1_1, test_y_1_1)
#     print("Train score: {}".format(model_1_1_train_score))
#     print("Test score: {}".format(model_1_1_test_score))
    print("Naive Bayes Multinomial model\n")
    print("Train accuracy:", np.round(accuracy_score(train_y_1_1, model_1_1.predict(train_x_1_1)), 2))
    print("Test accuracy:", np.round(accuracy_score(test_y_1_1, predict_y_1_1), 2))
    print("\n Confusion Matrix:\n", pd.crosstab(test_y_1_1, predict_y_1_1))
    print("\n Classification Report:\n",classification_report(test_y_1_1, predict_y_1_1))
#     return predict_y_1_1, model_1_1_train_score, model_1_1_test_score, model_1_1
    

def decision_tree_model(train_x_2, test_x_2, train_y_2, test_y_2):
    '''
    :param train_x_2: 80% of the train data from the observation variables
    :param test_x_2: 20% of the test data from the observation variables
    :param train_y_2: 80% of the train data from the response variables
    :param test_y_2: 20% of the test data from the response variables
    :return: Nil
    '''
    model_2 = DecisionTreeClassifier()
    model_2.fit(train_x_2, train_y_2)
    predict_y_2 = model_2.predict(test_x_2)
    model_2_train_score = model_2.score(train_x_2, train_y_2)
    model_2_test_score = model_2.score(test_x_2, test_y_2)
#     print("Train score: {}".format(model_2_train_score))
#     print("Test score: {}".format(model_2_test_score))
    print("Decision Tree model\n")
    print("Train accuracy:", np.round(accuracy_score(train_y_2, model_2.predict(train_x_2)), 2))
    print("Test accuracy:", np.round(accuracy_score(test_y_2, predict_y_2), 2),"\n")
    print("\n Confusion Matrix:\n", pd.crosstab(test_y_2, predict_y_2))
    print("\n Classification Report:\n",classification_report(test_y_2, predict_y_2))
#     return predict_y_2, model_2_train_score, model_2_test_score, model_2


def neural_network_model(train_x_3, test_x_3, train_y_3, test_y_3):
    '''
    :param train_x_3: 80% of the train data from the observation variables
    :param test_x_3: 20% of the test data from the observation variables
    :param train_y_3: 80% of the train data from the response variables
    :param test_y_3: 20% of the test data from the response variables
    :return: Nil
    '''
    model_3 = MLPClassifier(max_iter=2500, random_state=42)
    model_3.fit(train_x_3, train_y_3)
    predict_y_3 = model_3.predict(test_x_3)
    model_3_train_score = model_3.score(train_x_3, train_y_3)
    model_3_test_score = model_3.score(test_x_3, test_y_3)
#     print("Train score: {}".format(model_3_train_score))
#     print("Test score: {}".format(model_3_test_score))
    print("Neural network model\n")
    print("Train accuracy:", np.round(accuracy_score(train_y_3, model_3.predict(train_x_3)), 2))
    print("Test accuracy:", np.round(accuracy_score(test_y_3, predict_y_3), 2),"\n")
    print("\n Confusion Matrix:\n", pd.crosstab(test_y_3, predict_y_3))
    print("\n Classification Report:\n",classification_report(test_y_3, predict_y_3))
#     return predict_y_3, model_3_train_score, model_3_test_score, model_3

def random_forest_model(train_x_4, test_x_4, train_y_4, test_y_4):
    '''
    :param train_x_4: 80% of the train data from the observation variables
    :param test_x_4: 20% of the test data from the observation variables
    :param train_y_4: 80% of the train data from the response variables
    :param test_y_4: 20% of the test data from the response variables
    :return: Nil
    '''
    model_4 = RandomForestClassifier(random_state=42)
    model_4.fit(train_x_4, train_y_4)
    predict_y_4 = model_4.predict(test_x_4)
    model_4_train_score = model_4.score(train_x_4, train_y_4)
    model_4_test_score = model_4.score(test_x_4, test_y_4)
#     print("Train score: {}".format(model_4_train_score))
#     print("Test score: {}".format(model_4_test_score))
    print("Random Forest model\n")
    print("Train accuracy:", np.round(accuracy_score(train_y_4, model_4.predict(train_x_4)), 2))
    print("Test accuracy:", np.round(accuracy_score(test_y_4, predict_y_4), 2),"\n")
    print("\n Confusion Matrix:\n", pd.crosstab(test_y_4, predict_y_4))
    print("\n Classification Report:\n",classification_report(test_y_4, predict_y_4))
#     return predict_y_3, model_3_train_score, model_3_test_score, model_3

# def logistic_model(train_x_5, test_x_5, train_y_5, test_y_5):
#     '''
#     :param train_x_5: 80% of the train data from the observation variables
#     :param test_x_5: 20% of the test data from the observation variables
#     :param train_y_5: 80% of the train data from the response variables
#     :param test_y_5: 20% of the test data from the response variables
#     :return: Nil
#     '''
#     model_5 = LogisticRegression(random_state=42)
#     model_5.fit(train_x_5, train_y_5)
#     predict_y_5 = model_5.predict(test_x_5)
#     model_5_train_score = model_5.score(train_x_5, train_y_5)
#     model_5_test_score = model_5.score(test_x_5, test_y_5)
# #     print("Train score: {}".format(model_5_train_score))
# #     print("Test score: {}".format(model_5_test_score))
#     print("Train accuracy:", np.round(accuracy_score(train_y_5, model_5.predict(train_x_5)), 2))
#     print("Test accuracy:", np.round(accuracy_score(test_y_5, predict_y_5), 2),"\n")
#     print("\n Confusion Matrix:\n", pd.crosstab(test_y_5, predict_y_5))
#     print("\n Classification Report:\n",classification_report(test_y_5, predict_y_5))
```

```python
def classification_parser(dataset, classifier_model, reduction, component_count):
    x = dataset.drop(["STROKE", "RANDID", "AGE", "TIMEMI"], axis=1)
    if reduction:
        scaler = StandardScaler()
        scaler.fit(x)
        x = scaler.transform(x)
        pca = PCA(n_components=component_count)
        x = pca.fit_transform(x)
        x = pd.DataFrame(x)    
    y = dataset["STROKE"]
    if 2 in dataset["STROKE"]:
        print("At Risk\n")
    else:
        print("Healthy\n")
    # Splitting the data into train and test datasets
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
    # Running NaiveBayes Classification algorithm to check the model performance
    # predicted_value_1, train_score_1, test_score_1, model_type_1 = eval(classifier_model(X_train, X_test, y_train, y_test))
    eval(classifier_model+'(X_train, X_test, y_train, y_test)')
```

```python
df = pd.read_csv('Dataset_raw.csv')
print(df.shape)
```

```python
print(df.dtypes)
# Checking datatypes of the dataset for an categorical data
```

```python
# Checking missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Not dropping RANDID and TIME cause they do not contain any null value
# Data preprocessing ie, removing the missing values from the dataset
df = df.dropna(axis=0, how="any")
```

```python
# Rechecking the dataset to observe all invalid data are removed and that the dataset is solid
missing_values_remove_na = df.isnull().sum()
print(missing_values_remove_na)
```

```python
# Checking the shape of the new dataframe from the main dataframe shape of (620, 10)
print(df.shape)
```

```python
# Hence 24 dataset are removed from the main dataframe
# Converting the TOTCHOL column float dataset to categorical dataset
# Category for desirable cholesterol measurement is between 0 to 200
# Category for borderline high cholesterol measurement is between 200 to 239
# Category for High cholesterol measurement is between 240 and above
df = df.assign(TOTCHOL=pd.cut(df['TOTCHOL'], bins=[0, 199, 239, df['TOTCHOL'].max()], labels=['Desirable', 'Borderline', 'High']))
df.head()
```

```python
# Plot showing TOTCHOL vs STROKE
sns.countplot(x=df["TOTCHOL"], hue=df["STROKE"])
plt.title("TOTCHOL vs STROKE", fontsize = 14)
plt.show()
# The plot shows that Borderline individuals have highest chance of having a stroke
```

```python
# Converting the DIABP column float dataset to categorical dataset
# Category for normal measurement for diastolic blood pressure is between 0 to 90
# Category for borderline measurement for diastolic blood pressure is, when it is greater than 90
# and less than or equal to 100
# Category for high measurement for diastolic blood pressure is, when it is greater than 100
df = df.assign(DIABP=pd.cut(df['DIABP'], bins=[0, 90, 100, df['DIABP'].max()], labels=['Normal', 'Borderline', 'Hypertensive']))
df.head()
```

```python
# Plot showing DIABP vs STROKE
sns.countplot(x=df["DIABP"], hue=df["STROKE"])
plt.title("DIABP vs STROKE", fontsize = 14)
plt.show()
# The plot shows that the individuals high in diastolic blood pressure have highest chance of having a stroke
```

```python
# Converting the SYSBP column float dataset to categorical dataset
# Category for normal measurement for systolic blood pressure is between 0 to 140
# Category for borderline measurement for systolic blood pressure is, when it is greater than 140
# and less than or equal to 160
# Category for high measurement for systolic blood pressure is, when it is greater than 160
df = df.assign(SYSBP=pd.cut(df['SYSBP'], bins=[0, 140, 160, df['SYSBP'].max()], labels=['Normal', 'Borderline', 'Hypertensive']))
df.head()
```

```python
# Plot showing SYSBP vs STROKE
sns.countplot(x=df["SYSBP"], hue=df["STROKE"])
plt.title("SYSBP vs STROKE", fontsize = 14)
plt.show()
# The plot shows that the individuals high in systolic blood pressure have highest chance of having a stroke
```

```python
#Applying a condition to merge information from DIABP and SYSBP into one. That is BP
def conditions(dataframe):
    if (dataframe['DIABP'] > dataframe['SYSBP']) :
        return dataframe['DIABP']
    elif(dataframe['SYSBP'] > dataframe['DIABP']):
        return dataframe['SYSBP']
    elif(dataframe['SYSBP'] == dataframe['DIABP']):
        return dataframe['SYSBP']
```

```python
df['BP'] = df.apply(conditions, axis=1)
x = df.drop(['DIABP', 'SYSBP'], axis=1)
```
